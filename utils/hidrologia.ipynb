## Teste estatístico

```python
import numpy as np
from scipy.stats import kstest
from scipy.stats import genextreme
```

```python
print("\n" + "="*60)
print("TESTE DE KOLMOGOROV-SMIRNOV")
print("="*60)

dados = hmax1d['precipitacao máxima anual (mm)'].dropna().values
dados = dados[dados > 0]

ks_stat, ks_p = kstest(dados, 'genextreme', args=(c, loc, scale))

print(f"\nParâmetros GEV: c={c:.4f}, loc={loc:.2f}, scale={scale:.2f}")
print(f"KS statistic: {ks_stat:.4f}")
print(f"p-value: {ks_p:.4f}")

if ks_p < 0.05:
    print(f"\n→ REJEITA H₀: GEV não adequada (p={ks_p:.4f} < 0.05)")
else:
    print(f"\n→ NÃO REJEITA H₀: GEV adequada (p={ks_p:.4f} ≥ 0.05)")

# QQ-Plot rápido
plt.figure(figsize=(6, 5))
dados_ord = np.sort(dados)
n = len(dados_ord)
prob = np.arange(1, n + 1) / (n + 1)
q_teor = genextreme.ppf(prob, c, loc=loc, scale=scale)

plt.scatter(q_teor, dados_ord, alpha=0.7)
plt.plot([q_teor.min(), q_teor.max()], [q_teor.min(), q_teor.max()], 'r--')
plt.xlabel('Quantis Teóricos GEV')
plt.ylabel('Quantis Observados')
plt.title(f'QQ-Plot - {metadados["nome"]}')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# ANÁLISE DETALHADA DA NÃO ADERÊNCIA
# ============================================================

print("\n" + "="*60)
print("ANÁLISE DA NÃO ADERÊNCIA DA GEV")
print("="*60)

# 1. Teste KS já mostrou p-valor baixo
dados = hmax1d['precipitacao máxima anual (mm)'].dropna().values
dados = dados[dados > 0]
n = len(dados)

ks_stat, ks_p = kstest(dados, 'genextreme', args=(c, loc, scale))
print(f"Teste KS: D = {ks_stat:.4f}, p-valor = {ks_p:.4f}")

# 2. Calcula resíduos do QQ-Plot
dados_ord = np.sort(dados)
prob = np.arange(1, n + 1) / (n + 1)
q_teor = genextreme.ppf(prob, c, loc=loc, scale=scale)

residuos = dados_ord - q_teor
print(f"\nResíduos do QQ-Plot:")
print(f"  Média dos resíduos: {np.mean(residuos):.2f} mm")
print(f"  Desvio padrão dos resíduos: {np.std(residuos):.2f} mm")
print(f"  MAE (Erro Absoluto Médio): {np.mean(np.abs(residuos)):.2f} mm")

# 3. Identifica onde estão os maiores erros
print(f"\nAnálise por quartis:")
q1_idx = n // 4
q2_idx = n // 2
q3_idx = 3 * n // 4

print(f"  Q1 (25%): Erro médio = {np.mean(np.abs(residuos[:q1_idx])):.2f} mm")
print(f"  Q2 (50%): Erro médio = {np.mean(np.abs(residuos[q1_idx:q2_idx])):.2f} mm")
print(f"  Q3 (75%): Erro médio = {np.mean(np.abs(residuos[q2_idx:q3_idx])):.2f} mm")
print(f"  Q4 (100%): Erro médio = {np.mean(np.abs(residuos[q3_idx:])):.2f} mm")

# 4. Testa outras distribuições
print(f"\n" + "="*60)
print("TESTANDO OUTRAS DISTRIBUIÇÕES")
print("="*60)

from scipy.stats import gumbel_r, lognorm, weibull_min

distribuicoes = [
    ('GEV', 'genextreme', (c, loc, scale)),
]

# Ajusta e testa cada distribuição
resultados = []
for nome, dist_name, params in distribuicoes:
    if nome == 'GEV':
        ks_stat, ks_p = kstest(dados, dist_name, args=params)
        resultados.append((nome, ks_stat, ks_p, params))

# Testa Gumbel (especialmente para hidrologia)
loc_g, scale_g = gumbel_r.fit(dados)
ks_g, p_g = kstest(dados, 'gumbel_r', args=(loc_g, scale_g))
resultados.append(('Gumbel', ks_g, p_g, (loc_g, scale_g)))

# Testa Log-Normal
s, loc_ln, scale_ln = lognorm.fit(dados)
ks_ln, p_ln = kstest(dados, 'lognorm', args=(s, loc_ln, scale_ln))
resultados.append(('Log-Normal', ks_ln, p_ln, (s, loc_ln, scale_ln)))

# Testa Weibull
c_w, loc_w, scale_w = weibull_min.fit(dados)
ks_w, p_w = kstest(dados, 'weibull_min', args=(c_w, loc_w, scale_w))
resultados.append(('Weibull', ks_w, p_w, (c_w, loc_w, scale_w)))

# Ordena pelo p-valor (maior é melhor)
resultados.sort(key=lambda x: x[2], reverse=True)

print("\nComparação de distribuições (ordenado por p-valor):")
print("-"*80)
print(f"{'Distribuição':<12} {'KS D':<10} {'p-valor':<10} {'Adequada (α=0.05)':<15}")
print("-"*80)
for nome, ks, p, params in resultados:
    adequada = "SIM" if p >= 0.05 else "NÃO"
    print(f"{nome:<12} {ks:.6f}   {p:.6f}   {adequada:<15}")

# 5. Gráfico comparativo
print(f"\n" + "="*60)
print("GRÁFICO COMPARATIVO DAS DISTRIBUIÇÕES")
print("="*60)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle(f'Comparação de Distribuições - {metadados["nome"]}', fontsize=16)

# Dados para plots
x_range = np.linspace(dados.min() * 0.8, dados.max() * 1.2, 1000)

# A. Histograma com múltiplas distribuições
ax1 = axes[0, 0]
ax1.hist(dados, bins='auto', density=True, alpha=0.5, color='gray', label='Dados')
ax1.plot(x_range, genextreme.pdf(x_range, c, loc=loc, scale=scale), 
         'r-', linewidth=2, label='GEV')
ax1.plot(x_range, gumbel_r.pdf(x_range, loc_g, scale_g), 
         'b--', linewidth=2, label='Gumbel')
ax1.plot(x_range, lognorm.pdf(x_range, s, loc_ln, scale_ln), 
         'g-.', linewidth=2, label='Log-Normal')
ax1.set_xlabel('Precipitação (mm)')
ax1.set_ylabel('Densidade')
ax1.set_title('PDF das Distribuições')
ax1.legend()
ax1.grid(alpha=0.3)

# B. QQ-Plot da melhor distribuição
ax2 = axes[0, 1]
melhor_nome, melhor_ks, melhor_p, melhor_params = resultados[0]

if melhor_nome == 'GEV':
    q_teor_melhor = genextreme.ppf(prob, *melhor_params)
elif melhor_nome == 'Gumbel':
    q_teor_melhor = gumbel_r.ppf(prob, *melhor_params)
elif melhor_nome == 'Log-Normal':
    q_teor_melhor = lognorm.ppf(prob, *melhor_params)
else:
    q_teor_melhor = weibull_min.ppf(prob, *melhor_params)

ax2.scatter(q_teor_melhor, dados_ord, alpha=0.6)
ax2.plot([q_teor_melhor.min(), q_teor_melhor.max()], 
         [q_teor_melhor.min(), q_teor_melhor.max()], 'r--')
ax2.set_xlabel('Quantis Teóricos')
ax2.set_ylabel('Quantis Observados')
ax2.set_title(f'QQ-Plot: {melhor_nome} (p={melhor_p:.4f})')
ax2.grid(alpha=0.3)

# C. CDF comparativa
ax3 = axes[1, 0]
ecdf_x = np.sort(dados)
ecdf_y = np.arange(1, n + 1) / n
ax3.step(ecdf_x, ecdf_y, where='post', linewidth=2, label='ECDF Empírica')
ax3.plot(x_range, genextreme.cdf(x_range, c, loc=loc, scale=scale), 
         'r-', label='GEV CDF')
ax3.plot(x_range, gumbel_r.cdf(x_range, loc_g, scale_g), 
         'b--', label='Gumbel CDF')
ax3.set_xlabel('Precipitação (mm)')
ax3.set_ylabel('Probabilidade Acumulada')
ax3.set_title('CDF Comparativa')
ax3.legend()
ax3.grid(alpha=0.3)

# D. Resíduos por magnitude
ax4 = axes[1, 1]
residuos_melhor = dados_ord - q_teor_melhor
ax4.scatter(dados_ord, residuos_melhor, alpha=0.6)
ax4.axhline(y=0, color='r', linestyle='--')
ax4.set_xlabel('Precipitação Observada (mm)')
ax4.set_ylabel('Resíduos (Obs - Teórico)')
ax4.set_title(f'Resíduos: {melhor_nome}')
ax4.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(f"z_comparacao_distribuicoes_{metadados['nome']}.png", dpi=300)
plt.show()

# 6. Recomendações específicas
print(f"\n" + "="*60)
print("RECOMENDAÇÕES PARA SUA ANÁLISE")
print("="*60)

melhor_dist = resultados[0]
segunda_dist = resultados[1] if len(resultados) > 1 else None

print(f"\n1. DISTRIBUIÇÃO MAIS ADEQUADA: {melhor_dist[0]}")
print(f"   KS D = {melhor_dist[1]:.4f}, p-valor = {melhor_dist[2]:.4f}")

print(f"\n2. PROBLEMAS COM A GEV:")
print(f"   • QQ-Plot mostra desvios sistemáticos")
print(f"   • p-valor = {resultados[0][2]:.4f} (abaixo de 0.05)")
print(f"   • Provavelmente os dados têm assimetria diferente da GEV")

print(f"\n3. AÇÕES RECOMENDADAS:")
print(f"   a) Use {melhor_dist[0]} em vez de GEV")
print(f"   b) Verifique se há outliers influenciando o ajuste")
print(f"   c) Considere transformações nos dados (ex: log-transform)")

# 7. Verifica outliers
print(f"\n4. DIAGNÓSTICO DE OUTLIERS:")
Q1 = np.percentile(dados, 25)
Q3 = np.percentile(dados, 75)
IQR = Q3 - Q1
lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

outliers = dados[(dados < lim_inf) | (dados > lim_sup)]
print(f"   • Limites do boxplot: [{lim_inf:.1f}, {lim_sup:.1f}] mm")
print(f"   • {len(outliers)} valores outliers identificados")
if len(outliers) > 0:
    print(f"   • Valores outliers: {outliers}")
    print(f"   • Sugestão: Analise se são erros de medição ou eventos reais")

print(f"\n" + "="*60)
print("PRÓXIMOS PASSOS:")
print("="*60)
print("1. Recalcule as precipitações máximas usando {melhor_dist[0]}")
print("2. Gere novos gráficos com a distribuição adequada")
print("3. Atualize seu relatório com os novos resultados")
```
